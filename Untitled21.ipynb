**Removing null values and reseting indexing**

**Formatting the Trade_History for better analysis of data**

```python
df['Trade_History'] = df['Trade_History'].apply(ast.literal_eval)
```

**Making another dataframe to view the Trade_history**

```python
a=pd.DataFrame(df['Trade_History'][1])
a['time'] = pd.to_datetime(a['time'], unit='ms')
a.sort_values(by='time', inplace=True)
a['net_profit']=  a['realizedProfit'] + a['fee']
```

```python
length = len(df['Trade_History'])
```

**Calculating ROI**

```python
ROI = []
TOTAL_INVESTMENT = []
NET_PROFIT =[]
for i in range(length):
    try:
        # Convert the trade history to a DataFrame
        trade_history_df = pd.DataFrame(df['Trade_History'][i])

        # Calculate total investment
        total_investment = sum(trade_history_df['price'] * trade_history_df['quantity'])

        # Calculate net profit
        net_profit = sum(trade_history_df['realizedProfit'] + trade_history_df['fee'])

        # Calculate ROI
        roi = (net_profit / total_investment) * 100

        # Append the ROI to the results list
        TOTAL_INVESTMENT.append(total_investment)
        NET_PROFIT.append(net_profit)
        roi = round(roi, 4)
        ROI.append(roi)

    except Exception as e:
        print(f"Error processing row {i}: {e}")
        TOTAL_INVESTMENT.append(total_investment)
        NET_PROFIT.append(net_profit)
        ROI.append(None)  # Append None or some indication of error


```

```python
df['ROI'] =ROI
df['total_investment'] = TOTAL_INVESTMENT
df['net_profit'] = NET_PROFIT
```

**Calculating Sharp ratio**

```python
sharpRatio = []
for i in range(length):

    try:
        a = pd.DataFrame(df['Trade_History'][i])
        # Ensure 'time' is datetime and sort
        a['time'] = pd.to_datetime(a['time'], unit='ms')
        a.sort_values(by='time', inplace=True)

        # Calculate daily returns
        a['net_profit'] = a['realizedProfit'] + a['fee']
        daily_returns = a.groupby(a['time'].dt.date)['net_profit'].sum()
        average_daily_return = daily_returns.mean()
        std_daily_return = daily_returns.std()

        # Risk-free rate
        risk_free_rate = 0

        # Annualization factor for 90 days
        annualization_factor = np.sqrt(365 / 90)

        # Calculate Sharpe Ratio
        if std_daily_return == 0:
            print("Standard deviation of returns is zero. Cannot calculate Sharpe Ratio.")
            sharpe_ratio = None
        else:
            sharpe_ratio = (average_daily_return - risk_free_rate) / std_daily_return * annualization_factor
            sharpe_ratio = round(sharpe_ratio, 4)

        sharpRatio.append(sharpe_ratio)

    except Exception as e:
        print(f"Error processing row {i}: {e}")
        sharpRatio.append(None)
df['Sharp_Ratio'] = sharpRatio
```

**Calculating maximum dropdown**

```python
mdd = []
for i in range(length):

  try:
    a = pd.DataFrame(df['Trade_History'][i])
    a['time'] = pd.to_datetime(a['time'], unit='ms')
    a.sort_values(by='time', inplace=True)
    initial_balance = 100_000
    a['net_profit'] = a['realizedProfit'] + a['fee']
    # Calculate the running balance
    a['balance'] = initial_balance + a['net_profit'].cumsum()

    # Calculate running maximum balance
    a['running_max'] = a['balance'].cummax()

    # Calculate drawdown
    a['drawdown'] = a['running_max'] - a['balance']

    # Calculate maximum drawdown
    max_drawdown = a['drawdown'].max()

    # Peak balance
    peak_balance = a['running_max'].max()

    # Calculate MDD percentage
    mdd_percentage = (max_drawdown / peak_balance) * 100
    mdd_percentage = round(mdd_percentage, 4)
    mdd.append(mdd_percentage)
  except Exception as e:
    print(f"Error processing row {i}: {e}")
    mdd.append(None)  # Append None or some indication of error

df['MDD'] = mdd
```

**Calculating winning Trade**

```python
winning_trades = []

# Iterate through the length of 'Trade_History'
for i in range(length):
    x = 0
    try:
        # Create a DataFrame for the current row
        trade_history_df = pd.DataFrame(df['Trade_History'][i])

        # Check if any realizedProfit is greater than 0
        trade_history_df['net_profit'] = trade_history_df['realizedProfit'] + trade_history_df['fee']
        for profit in trade_history_df['net_profit']:
            if profit > 0:
                x += 1

        winning_trades.append(x)

    except Exception as e:
        print(f"Error processing row {i}: {e}")
        winning_trades.append(None)  # Append None or some indication of error

print(f"Number of winning trades: {winning_trades}")

df['winning_trades'] = winning_trades
```

```python
# Initialize list to store the number of trades
noOfTrades = []

# Iterate through the length of 'Trade_History'
for i in range(len(df['Trade_History'])):
    try:
        # Extract the trade history for the current row
        a = df['Trade_History'][i]

        # Check if a is a list or DataFrame and append its length
        if isinstance(a, list) or isinstance(a, pd.DataFrame):
            noOfTrades.append(len(a))
        else:
            raise ValueError(f"Unexpected data type in row {i}: {type(a)}")

    except Exception as e:
        print(f"Error processing row {i}: {e}")
        noOfTrades.append(None)  # Append None or some indication of error

df['noOfTrades'] = noOfTrades
```

```python
df.drop(columns=['Trade_History'], inplace=True)
```

```python
df['winning_ratio']= df['winning_trades']/df['noOfTrades']*100
```

```python
scaler = MinMaxScaler()
metrics = ['noOfTrades', 'winning_trades', 'winning_ratio', 'ROI', 'total_investment', 'net_profit', 'MDD', 'Sharp_Ratio']
df[metrics] = scaler.fit_transform(df[metrics])
```

```python
weights = {
    'noOfTrades': 0.1,
    'winning_trades': 0.1,
    'winning_ratio': 0.1,
    'ROI': 0.2,
    'total_investment': 0.1,
    'net_profit': 0.2,
    'MDD': -0.1,  # Negative weight for drawdown as lower drawdown is better
    'Sharp_Ratio': 0.2
}
```

```python
df['Overall_Score'] = (
    df['noOfTrades'] * weights['noOfTrades'] +
    df['winning_trades'] * weights['winning_trades'] +
    df['winning_ratio'] * weights['winning_ratio'] +
    df['ROI'] * weights['ROI'] +
    df['total_investment'] * weights['total_investment'] +
    df['net_profit'] * weights['net_profit'] +
    df['MDD'] * weights['MDD'] +
    df['Sharp_Ratio'] * weights['Sharp_Ratio']
)

# 4. Rank accounts based on overall score
df['Rank'] = df['Overall_Score'].rank(ascending=False)

# Select top 20 accounts (assuming there are more accounts in the full dataset)
top_20_accounts = df.nlargest(20, 'Overall_Score')

# Save results to CSV
top_20_accounts.to_csv('top_20_accounts.csv', index=False)
```
